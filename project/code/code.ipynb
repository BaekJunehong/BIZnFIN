```python
import pandas as pd

# 열 이름 변환 함수
def transform_column_name(col_name):
    parts = col_name.split('\n')
    if len(parts) == 2:
        year, desc = parts
        desc = desc.split('.')[1]  # 변수 설명 부분만 추출
        return f"{year}/{desc}"
    return col_name

# 열 이름 변환 적용
data.columns = [transform_column_name(col) for col in data.columns]

data.head()
```
